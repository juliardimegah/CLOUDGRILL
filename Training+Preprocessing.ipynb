```python
crot6 = {
  'pantai' : 1,
 'kota' : 2,
 'pegunungan' :  3,
 'gunung' : 4

}
crot3 = {
    'panas' : 0,
    'hujan' : 1 
}
num_code = 1
data_tokenize = []
text_to_num = {}

for index, rows in df.iterrows():
  new_tokenize_barang_bawaan = []
  stack_data_tokenize = []
  for bawaan in rows['barang_bawaaan']:
    if bawaan == '':
      continue
    
    if bawaan in text_to_num:
      new_tokenize_barang_bawaan.append(text_to_num.get(bawaan))
    else:
      text_to_num[bawaan] = num_code
      num_code += 1
      new_tokenize_barang_bawaan.append(text_to_num.get(bawaan))

  stack_data_tokenize.append(crot6.get(rows['lokasi']))
  stack_data_tokenize.append(crot3.get(rows['cuaca']))
  stack_data_tokenize.append(new_tokenize_barang_bawaan)

  data_tokenize.append(stack_data_tokenize)
# If the number is positive, we print an appropriate message
```

```python
 import numpy as np
 from keras.preprocessing import sequence
 from keras.models import Sequential
 from keras.layers import Dense, Dropout, Embedding, LSTM, Bidirectional
 from keras.datasets import imdb 
 n_unique_words = 10000 # cut texts after this number of words
 maxlen = 200
 batch_size = 128 
```

```python
(x_train, y_train),(x_test, y_test) = imdb.load_data(num_words=n_unique_words)
```

```python
 x_train = sequence.pad_sequences(x_train, maxlen=maxlen)
 x_test = sequence.pad_sequences(x_test, maxlen=maxlen)
 y_train = np.array(y_train)
 y_test = np.array(y_test)  
 model = Sequential()
 model.add(Embedding(n_unique_words, 128, input_length=maxlen))
 model.add(Bidirectional(LSTM(64)))
 model.add(Dropout(0.5))
 model.add(Dense(1, activation='sigmoid'))
 model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) 
```

```python
 history=model.fit(x_train, y_train,
           batch_size=batch_size,
           epochs=12,
           validation_data=[x_test, y_test])
 
 print(history.history['cuaca']) 
 print(history.history['lokasi'])
```

```python
 from matplotlib import pyplot
 pyplot.plot(history.history['lokasi'])
 pyplot.plot(history.history['cuaca'])
 pyplot.title('model loss vs accuracy')
 pyplot.xlabel('epoch')
 pyplot.legend(['loss', 'lokasi'], loc='upper right')
 pyplot.show() 
```
